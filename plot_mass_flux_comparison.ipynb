```python
import xarray as xr
import proplot as pplt
import pandas as pd
import numpy as np
from pathlib import Path

from AOSCMcoupling.files import OIFSPreprocessor, NEMOPreprocessor
```

```python
start_date = pd.Timestamp("2014-07-01")
oifs_preprocessor = OIFSPreprocessor(
    start_date, np.timedelta64(-7, "h")
)
nemo_preprocessor = NEMOPreprocessor(
    start_date, np.timedelta64(-7, "h")
)
```

```python
ece3_dir = Path("PAPA")
```

```python
def load_progvar(output_dir, exp_ids):
    oifs_progvar = [
        xr.open_mfdataset(
            f"{output_dir}/{exp_id}/progvar.nc", preprocess=oifs_preprocessor.preprocess
        )
        for exp_id in exp_ids
    ]

    return oifs_progvar
```

```python
exp_ids = ["CEX0", "CEX1", "CEX2", "CEXS_30"]
oifs_progvars = load_progvar(Path("PAPA"), exp_ids)
```

```python
exp_ids = ["MFN0", "MFN1", "MFN2", "MFNS_30"]
oifs_progvars_mfn = load_progvar(Path("PAPA"), exp_ids)
```

```python
colors = ["m", "c", "y", "k"]
labels = ["parallel", "atm-first", "oce-first", "converged SWR"]
alpha = 1
linestyles = ["--", ":", "-.", "-"]
```
